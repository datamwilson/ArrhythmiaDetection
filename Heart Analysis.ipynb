```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.utils import shuffle
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/arrhythmia/arrhythmia.data', header=None)
```

```python
df = shuffle(df)
```

```python
df.info()
```

```python
df.head()
```

```python
labels = df[279]
```

```python
data = df
data.drop(columns=[279])
```

```python
for column in data.columns.values:
    data[column].replace('?',0,inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.30, random_state=0)
```

```python
logisticRegr = LogisticRegression()
```

```python
logisticRegr.fit(x_train, y_train)
```

```python
score = logisticRegr.score(x_test, y_test)
```
